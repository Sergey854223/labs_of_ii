```python
# импортируем библиотеку
import pandas as pd
# применим функцию read_csv() и посмотрим на первые три записи файла
train = pd.read_csv('/content/Marvel Movies.csv')
train.head(3)
```

```python
train.info()
```

```python
# для построения графиков воспользуемся новой для нас библиотекой seaborn
import seaborn as sns
```

```python
# посмотрим кол-во живых и мертвый персонажей по гендеру
# с помощью x и hue мы можем уместить две категориальные переменные на одном графике
sns.countplot(x = 'SEX', hue = 'ALIVE', data = train)
```

```python
# выявим пропущенные значения с помощью .isnull() и посчитаем их количество sum()
train.isnull().sum()
```

```python
# переменная GSM (сексуальная группа), скорее всего, не является самой важной
# избавимся от нее с помощью метода .drop()
# (параметр axis отвечает за столбцы, inplace = True сохраняет изменения)
train.drop(columns = 'GSM', axis = 1, inplace = True)
```

```python
# а вот APPEARANCES  (появление) точно важен, заменим пустые значения средним арифметическим
train['APPEARANCES'].fillna(train['APPEARANCES'].mean(), inplace = True)
```

```python
# у нас остаются две пустые строки в ALIVE, удалим их
train.dropna(inplace = True)
```

```python
train.isnull().sum()
```

```python
# применим one-hot encoding к переменной Sex (пол) спомощью метода .get_dummies()
pd.get_dummies(train['SEX']).head(3)
```

```python
# удалим первый столбец, он избыточен
sex = pd.get_dummies(train['SEX'], drop_first = True)
sex.head(3)
```

```python
eye = pd.get_dummies(train['EYE'], drop_first = True)
hair = pd.get_dummies(train['HAIR'], drop_first = True)
```

```python
train = pd.concat([train, eye, sex, hair], axis = 1)
```

```python
# применим функцию drop() к соответствующим столбцам
train.drop(['EYE', 'HAIR', 'APPEARANCES', 'ALIGN'], axis = 1, inplace = True)
train.head(3)
```

```python
# импортируем класс StandardScaler
from sklearn.preprocessing import StandardScaler
# создадим объект этого класса
scaler = StandardScaler()
# выберем те столбцы, которые мы хотим масштабировать
cols_to_scale = ['APPEARANCE', 'Year']
# рассчитаем среднее арифметическое и СКО для масштабирования данных
scaler.fit(train[cols_to_scale])
# применим их
train[cols_to_scale] = scaler.transform(train[cols_to_scale])
# посмотрим на результат
train.head(3)
```

```python
# импортируем библиотеку
import pandas as pd
# применим функцию read_csv() и посмотрим на первые три записи файла
train6 = pd.read_csv('/content/Marvel Movies.csv')
train6.head(3)
```

```python
# переменная GSM (сексуальная группа), скорее всего, не является самой важной
# избавимся от нее с помощью метода .drop()
# (параметр axis отвечает за столбцы, inplace = True сохраняет изменения)
train6.drop(columns = 'GSM', axis = 1, inplace = True)
```

```python
# а вот APPEARANCES  (появление) точно важен, заменим пустые значения средним арифметическим
train6['APPEARANCES'].fillna(train6['APPEARANCES'].mean(), inplace = True)
```

```python
# у нас остаются две пустые строки в ALIVE, удалим их
train6.dropna(inplace = True)
```

```python
train6.isnull().sum()
```

```python
# применим one-hot encoding к переменной Sex (пол) спомощью метода .get_dummies()
pd.get_dummies(train6['SEX']).head(6)
```

```python
# удалим первый столбец, он избыточен
sex = pd.get_dummies(train6['SEX'], drop_first = True)
sex.head(6)
```

```python
eye = pd.get_dummies(train6['EYE'], drop_first = True)
hair = pd.get_dummies(train6['HAIR'], drop_first = True)
```

```python
train6 = pd.concat([train6, eye, sex, hair], axis = 1)
```

```python
# применим функцию drop() к соответствующим столбцам
train6.drop(['EYE', 'HAIR', 'FIRST APPEARANCE', 'ALIGN'], axis = 1, inplace = True)
train6.head(3)
```

```python
# импортируем класс StandardScaler
from sklearn.preprocessing import StandardScaler
# создадим объект этого класса
scaler = StandardScaler()
# выберем те столбцы, которые мы хотим масштабировать
cols_to_scale = ['APPEARANCES', 'Year']
# рассчитаем среднее арифметическое и СКО для масштабирования данных
scaler.fit(train6[cols_to_scale])
# применим их
train6[cols_to_scale] = scaler.transform(train6[cols_to_scale])
# посмотрим на результат
train6.head(3)
```

```python
train6.columns
```

```python
train6.columns = train6.columns.map(str)
```

```python
# поместим в X_train все кроме столбца ALIVE
X_train = train6.drop(['name', 'ID', 'SEX', 'ALIVE'], axis = 1)
# столбец 'ALIVE' станет нашей целевой переменной (y_train)
y_train = train6['ALIVE']
X_train.head(3)
```

```python
# импортируем логистическую регрессию из модуля linear_model библиотеки sklearn
from sklearn.linear_model import LogisticRegression
# создадим объект этого класса и запишем его в переменную model
model = LogisticRegression()
# обучим нашу модель
model.fit(X_train, y_train)
```

```python
# сделаем предсказание класса на обучающей выборке
y_pred_train = model.predict(X_train)
```

```python
# построим матрицу ошибок
from sklearn.metrics import confusion_matrix
# передадим ей фактические и прогнозные значения
conf_matrix = confusion_matrix(y_train, y_pred_train)
# преобразуем в датафрейм
conf_matrix_df = pd.DataFrame(conf_matrix)
conf_matrix_df
```

```python
conf_matrix_labels = pd.DataFrame(conf_matrix, columns =['Прогноз_мертвый', 'Прогноз_живой'], index = ['Факт_мертвый', 'Факт_живой'])
conf_matrix_labels
```
